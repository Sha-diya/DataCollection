```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

# Extract data of weather


```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.meteoprog.com/weather/Dhaka/month/',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
weatherData=soup.find_all('div','region')
```

```python
weatherData
```

```python
city=[]
temp=[]

for i in weatherData:
  city.append(i.find('a','region__name').text)
  temp.append(i.find('div','region__temperature').text.strip())
```

```python
len(temp)
```

```python
finalData=pd.DataFrame({'city':city,'temp':temp})
```

```python
finalData
```

```python
finalData.to_csv("WeatherData.csv",index=False)
```
