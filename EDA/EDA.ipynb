<h2>Bivariate Analysis + Multivariate Analysis</h2>

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris=sns.load_dataset('iris')
```

```python
tips.head()
```

1. ScatterPlot --(Numerical - Numerical)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

2. BarPlot (Neumerical - categorical)

```python
df.head()
```

```python
sns.barplot(y=df['stories'],x=df['prefarea'],hue=df['mainroad'])
```

3. BoxPlot (Numerical - categorical)

```python
sns.boxplot(x=df['prefarea'],y=df['price'],hue=df['furnishingstatus'])
```

4. DistPlot (Numerical - categorical)

```python
sns.displot(df[df['area']>7000]['price'],kde=True)
sns.displot(df[df['area']<=7000]['price'],kde=True)
```

5. HeatMap (Categorical - Categorical)

```python
df.head()
```

```python
sns.heatmap(pd.crosstab(df['mainroad'],df['stories']))
```

```python
flights.head()
```

```python
flights.groupby('month')['passengers'].mean()
```

```python
df.head()
```

```python
sns.clustermap(pd.crosstab(df['guestroom'],df['prefarea']))
```

PairPlot

```python
iris.head()
```

LinePLot (Neumerical - neumerical)
X-axis time based

```python
flights.head()
```

```python
new = flights.groupby('year').sum(numeric_only=True).reset_index()
```

```python
sns.lineplot(x='year',y='passengers', data=new)
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
